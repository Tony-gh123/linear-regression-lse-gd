```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
data = pd.read_csv('data/hsb2.csv')

# Split into features (x) and target (y)
x = data["math"].values
y = data["science"].values
```

```python
print(f"Total {len(x)} samples")
print(f"math:  min={x.min():.1f}, max={x.max():.1f}, mean={x.mean():.1f}, std={x.std():.1f}, total={len(x):.1f}")
print(f"science: min={y.min():.1f}, max={y.max():.1f}, mean={y.mean():.1f}, std={y.std():.1f}, total={len(y):.1f}")

# --- Quick checks ---
print(f"\nMissing-values: math: {np.isnan(x).sum()}, science: {np.isnan(y).sum()}")
print(f"Correlation: {np.corrcoef(x, y)[0,1]:.3f}")

# --- Plot ---
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.scatter(x, y, alpha=0.6, color='teal')
plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.title(f"Scatter (r = {np.corrcoef(x, y)[0,1]:.3f})")
plt.grid(True, alpha=0.3)

plt.subplot(1,2,2)
plt.hist(x, bins=15, alpha=0.7, label='Math', color='skyblue')
plt.hist(y, bins=15, alpha=0.7, label='Science', color='salmon')
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.legend()
plt.title("Distributions")
plt.tight_layout()
plt.show()

print("\nReady for modeling!")
```
